```python
import numpy as np
# Load EM traces and inputs
em_traces = np.load('avg_traces.npy')
inputs = np.load('avg_entries.npy')
```

```python
# Calculate the average of all samples in the subset of traces
#Plotting all these points shows changes in the average of the traces
#over time and can help you find anomalies in your trace acquisition
#campaign; 

averages_per_trace = np.mean(em_traces, axis=1)

# Plotting these points to show changes in the average of the traces over time
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(averages_per_trace* 1e5, '-o', markersize=5)
plt.title("Average Value of Each Trace Over Time")
plt.xlabel("Trace Number (Em-trace)")
plt.ylabel("Average Value (10µ Volt)")
plt.grid(True)
plt.show()
```

```python
# Calculate the average and standard deviation per sample

averages_per_sample = np.mean(em_traces, axis=0)
std_dev_per_sample = np.std(em_traces, axis=0)

# Plotting the average and standard deviation per sample

plt.figure(figsize=(12, 6))
plt.plot(averages_per_sample * 1e2, label="Average", color="blue")  # Multiply by 1e5 to convert to 10µ Volt units
plt.fill_between(range(len(averages_per_sample)), 
                 (averages_per_sample - std_dev_per_sample) * 1e2, 
                 (averages_per_sample + std_dev_per_sample) * 1e2, 
                 color="lightblue", label="Standard Deviation")
plt.title("Average and Standard Deviation Value Per Sample")
plt.xlabel("Sample Number")
plt.ylabel("Value (10m Volt)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting both raw traces (top) and sample-averaged trace (bottom)

fig, ax = plt.subplots(2, 1, figsize=(12, 10), sharex=True)

# Plot raw traces (top)
for trace in em_traces:
    ax[0].plot(trace * 1e5, color='lightgray')  # Multiply by 1e5 to convert to 10µ Volt units
ax[0].set_title("Raw Traces")
ax[0].set_ylabel("Value (10µ Volt)")
ax[0].grid(True)

# Plot sample-averaged trace (bottom)
ax[1].plot(averages_per_sample * 1e5, label="Average", color="blue")  # Multiply by 1e5 to convert to 10µ Volt units
ax[1].fill_between(range(len(averages_per_sample)), 
                   (averages_per_sample - std_dev_per_sample) * 1e5, 
                   (averages_per_sample + std_dev_per_sample) * 1e5, 
                   color="lightblue", label="Standard Deviation")
ax[1].set_title("Sample-Averaged Trace")
ax[1].set_xlabel("Sample Number")
ax[1].set_ylabel("Value (10µ Volt)")
ax[1].legend()
ax[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
# Convert the floating point number to IEEE 754 format

def float_to_ieee_754(n):
    # 1. Determine the sign bit
    if n < 0:
        sign = 1
        n = -n
    else:
        sign = 0
    
    # 2. Convert the number into binary
    binary_representation = ''
    while n > 0 and len(binary_representation) <= 24:
        n = n * 2
        binary_representation += str(int(n))
        n = n - int(n)
    
    # 3. Normalize the binary number
    # In this case, the number is already normalized as it is < 1 and starts with 0.
    
    # 4. Determine the exponent and bias it
    # Since the number is in form 0.xxxxx, the exponent is -1
    exponent = -1 + 127  # Adding the bias
    
    # 5. Determine the fraction (mantissa)
    fraction = binary_representation[1:24]  # Omit the leading 1 and consider only the next 23 bits
    
    # 6. Combine the results
    ieee_754_representation = f"{sign:01b}{exponent:08b}{fraction}"
    return ieee_754_representation

number = 0.793281
ieee_754_format = float_to_ieee_754(number)
ieee_754_format
```

```python
# Redefining the ieee_754_to_float function

def ieee_754_to_float(sign, exponent, fraction):
    # Convert biased exponent to true exponent
    true_exponent = int(exponent, 2) - 127
    
    # Convert fraction to decimal
    mantissa_decimal = 0
    for i, bit in enumerate(fraction, start=1):
        mantissa_decimal += int(bit) * (2 ** -i)
    
    # Calculate the float value
    float_value = (-1)**sign * (1 + mantissa_decimal) * (2 ** true_exponent)
    return float_value

# Generate all 16-bit combinations
hypotheses = []
for exponent in range(256):  # All 8-bit combinations
    for fraction in range(256):  # First 8 bits of mantissa
        full_exponent = f"{exponent:08b}"
        full_fraction = f"{fraction:08b}" + "0" * 15  # Adding 15 zeros to make it 23 bits
        hypotheses.append(ieee_754_to_float(0, full_exponent, full_fraction))

# Filter the hypotheses based on the given range

C = 2.5
d0 = 5

filtered_hypotheses = [value for value in hypotheses if 0 <= value <= 5]

filtered_hypotheses[:10], len(filtered_hypotheses)  # Display the first 10 filtered values and the total number of filtered values
min_value = min(filtered_hypotheses)
max_value = max(filtered_hypotheses)
mean_value = sum(filtered_hypotheses) / len(filtered_hypotheses)
shape_of_weights = len(filtered_hypotheses)

min_value, max_value, mean_value,shape_of_weights
```

```python
# ------------------------ Hamming Weight functions ------------------------ #

_HW8_table = np.array([bin(x).count('1') for x in range(256)], dtype = np.uint8)

def hamming_weight_8(x):
    return _HW8_table[x]

def hamming_weight_16(x):
    return hamming_weight_8(x & 0xFF) + hamming_weight_8(x >> 8)

def hamming_weight_32(x):
    return hamming_weight_16(x & 0xFFFF) + hamming_weight_16(x >> 16)

def hamming_weight_64(x):
    return hamming_weight_32(x & 0xFFFFFFFF) + hamming_weight_32(x >> 32)

def HW_IEEE754(x):
    return hamming_weight_32(x)

def local_HW_IEEE754(x, decalage):
    return hamming_weight_32(x >> decalage)

def hamming_weight(x):
    return bin(x).count('1')

def get_hamming_weight(a, b):
    c = a * b
    tmp = np.array([c], dtype=np.float32)
    tmp_int32 = tmp.view(np.int32)
    HW_tmp = HW_IEEE754(tmp_int32).astype(np.uint8)
    return HW_tmp[0]
```

```python
def generate_one_sim_traces(input, secret_value):
    nb_sample = 10000
    trace = np.zeros((nb_sample,))
    for i in range(nb_sample):
        trace[i] = np.random.randint(1,3)

    # Assign leaking point related to MAC operations
    trace[1] = get_hamming_weight(input, secret_value)
    trace = apply_Gaussian_noise(np.array(trace, dtype=np.float32), mean=0, std=5)
    return trace
```

```python
def apply_Gaussian_noise(traces, mean=0, std=0.2):
    noise = np.random.normal(loc = mean, scale = std, size=traces.shape)
    new_traces = traces + noise
    return new_traces.astype(dtype = np.float32)
```

```python
import numpy as np

def sim_mul_to_bitwise(input_val, secret_weight):
    product = input_val * secret_weight
    
    # Convert the product to a 32-bit float and then view it as a 32-bit integer
    product_int32 = np.float32(product).view(np.int32)
    
    # Convert the integer to its binary representation and return
    return bin(product_int32)

def extract_most_significant_bit(bitwise_output):
    """
    Extract the most significant bit after the sign bit from the bitwise output.
    
    :param bitwise_output: The binary representation of the number.
    :return: Integer value (0 or 1) representing the extracted bit.
    """
    # If the bitwise_output is too short, return 0
    if len(bitwise_output) <= 3:
        return 0
    
    # Otherwise, extract the bit after the sign bit
    return int(bitwise_output[3])

def compute_most_significant_leakage(input_val, secret_weight):
    bitwise_output = sim_mul_to_bitwise(input_val, secret_weight)
    return extract_most_significant_bit(bitwise_output)

# Test the functions with a sample input
sample_input = 1.5
sample_secret_weight = 0.793281

bitwise_output_sample = sim_mul_to_bitwise(sample_input, sample_secret_weight)
most_significant_leakage_sample = compute_most_significant_leakage(sample_input, sample_secret_weight)

bitwise_output_sample, most_significant_leakage_sample

```

```python
def num_same(a, b):
    
    if len(a) != len(b):
        raise ValueError("Arrays must be same length!")

    if max(a) != max(b):
        raise ValueError("Arrays max() should be the same!")
    
    #Count how many list items match up
    same = 0
    for i, _ in enumerate(a):
        if a[i] == b[i]:
            same += 1
    
    return same

# Re-run the adapted guessing loop
results = []
for guess in filtered_hypotheses:
    
    # Get a hypothetical leakage list using the flattened version of inputs
    hypothetical_leakage = [compute_most_significant_leakage(input_val, guess) for input_val in flattened_inputs]
    
    # Use the num_same function
    same_count = num_same(hypothetical_leakage, most_significant_leakage_data)
    
    # Store the results
    results.append((guess, same_count))

results

```

```python
# Initialize an empty list of size equal to the length of filtered_hypotheses to store the match counts for each guess
guess_list = [0] * len(filtered_hypotheses)

# Iterate over each guess and compute the match count
for idx, guess in enumerate(filtered_hypotheses):
    
    # Get a hypothetical leakage list using the flattened version of inputs
    hypothetical_leakage = [compute_most_significant_leakage(input_val, guess) for input_val in flattened_inputs]
    
    # Use the num_same function
    same_count = num_same(hypothetical_leakage, most_significant_leakage_data)
    
    # Store the count in the guess_list at the appropriate index
    guess_list[idx] = same_count

# Use np.argsort to generate a list of indices from high to low match counts
sorted_list = np.argsort(guess_list)[::-1]

# Print top 5 only
top_5_results = []
for index in sorted_list[0:5]:
    formatted_result = ("Key Guess {:02X} = {:04d} matches".format(int(filtered_hypotheses[index]*100), guess_list[index]))
    top_5_results.append(formatted_result)

top_5_results
```

```python
def float_to_bytes(float_val):
    # Convert the float to its 32-bit integer representation
    int_repr = np.float32(float_val).view(np.int32)
    
    # Convert the integer to its 32-bit binary representation
    bin_repr = format(int_repr, '032b')
    
    # Extract the 8 bits for the exponent
    exponent = bin_repr[1:9]
    
    # Extract the most significant 8 bits of the mantissa
    mantissa_msb = bin_repr[9:17]
    
    # Convert the binary strings to bytes
    exponent_byte = int(exponent, 2).to_bytes(1, 'big')
    mantissa_byte = int(mantissa_msb, 2).to_bytes(1, 'big')
    
    return exponent_byte, mantissa_byte

# Re-process the top_5_results with the conversion function
top_5_converted = []

for result in top_5_results:
    # Extracting the integer value (key guess) from the result string
    key_guess_int = int(result.split(" ")[2], 16)  # Convert hex string to integer
    
    # Convert integer to its floating point representation (revert the earlier multiplication by 100)
    key_guess_float = key_guess_int / 100.0
    
    exponent_byte, mantissa_byte = float_to_bytes(key_guess_float)
    exponent_hex = "0x{:02X}".format(exponent_byte[0])
    mantissa_hex = "0x{:02X}".format(mantissa_byte[0])
    
    # Formatting the result with the hexadecimal representation
    formatted_result = "{} - Exponent: {}, Mantissa MSB: {}".format(result, exponent_hex, mantissa_hex)
    top_5_converted.append(formatted_result)

top_5_converted
```
